```python
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb

# -----------------------------
# FEATURES (NO COUNTRY)
# -----------------------------
features = [
    "views", "likes", "comments", "Engagement_Rate",
    "watch_time_minutes", "video_length_minutes", "subscribers"
]

# -----------------------------
# PREPROCESSOR (SCALING)
# -----------------------------
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), features)
    ]
)

# -----------------------------
# XGBOOST MONOTONIC CONSTRAINTS
# (+1 means feature must INCREASE revenue)
# -----------------------------
constraints = "(1,1,1,1,1,1,1)"
# One '1' for each feature

xgb_model = xgb.XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    monotone_constraints=constraints,
    objective='reg:squarederror',
    random_state=42
)

# -----------------------------
# FINAL PIPELINE
# -----------------------------
pipeline = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", xgb_model)
])

# -----------------------------
# TRAIN MODEL
# -----------------------------
pipeline.fit(x_train, y_train)

# -----------------------------
# EVALUATE MODEL
# -----------------------------
pred = pipeline.predict(x_test)

pred_real = np.expm1(pred)    # convert log values back to real revenue
y_real = np.expm1(y_test)

print("XGBoost Monotonic Model (Final)")
print("MAE :", mean_absolute_error(y_real, pred_real))
print("MSE :", mean_squared_error(y_real, pred_real))
print("R2  :", r2_score(y_real, pred_real))

# -----------------------------
# SAVE FINAL MODEL
# -----------------------------
joblib.dump(pipeline, "pipeline_xgb_monotonic.joblib")
print("Model saved as pipeline_xgb_monotonic.joblib")
```
